##Q5 is_prime_member, is_prime_eligible, is_festival_sale

```python
df = pd.read_csv(file_path)

# Define a universal boolean cleaner
def clean_boolean(value):
    try:
        value = str(value).strip().lower()
        if value in ['true', 'yes', 'y', '1']:
            return True
        elif value in ['false', 'no', 'n', '0']:
            return False
        else:
            return False  # or use pd.NA if you want to preserve missing
    except:
        return False

# Apply to all relevant columns
boolean_columns = ['is_prime_member', 'is_prime_eligible', 'is_festival_sale']
for col in boolean_columns:
    df[col] = df[col].apply(clean_boolean)
    
df['is_prime_member']=df['is_prime_member'].astype(bool)
df['is_prime_eligible']=df['is_prime_eligible'].astype(bool)
df['is_festival_sale']=df['is_festival_sale'].astype(bool)
# Save the cleaned data back to the same file
df.to_csv(file_path, index=False)

print(" Boolean columns standardized to True/False and file updated successfully.")


  
```

```python

import os
# Q9: Fix outlier prices


df = pd.read_csv(file_path)

# Normalize column names
df.columns = df.columns.str.strip().str.lower().str.replace('₹', '').str.replace(' ', '_')

# IQR bounds for fallback correction
Q1 = df['original_price_inr'].quantile(0.25)
Q3 = df['original_price_inr'].quantile(0.75)
IQR = Q3 - Q1
upper_bound = Q3 + 1.5 * IQR

# Define expected price ranges by subcategory
subcategory_price_ranges = {
    'smartphones': (5000, 100000),
    'smart_watch': (500, 15000),
    'audio': (300, 20000),
    'tablets': (5000, 60000),
    'laptops': (20000, 150000),
    'tv_&_entertainment': (10000, 150000)
}

# Correction logic using subcategory
def correct_price(row):
    price = row['original_price_inr']
    subcat = str(row.get('subcategory', '')).strip().lower().replace(' ', '_')
    expected = subcategory_price_ranges.get(subcat)
    
    if expected:
        min_p, max_p = expected
        if price > max_p * 10:
            return price / 100
        elif price < min_p / 10:
            return price * 100
    if price > upper_bound:
        return price / 100
    return price

# Apply correction
df['original_price_inr'] = df.apply(correct_price, axis=1)

# Save updated file
df.to_csv(file_path, index=False)
print("✅ Outlier prices corrected and saved for amazon_india_2024.csv")
```

```python
import os
import pandas as pd
import numpy as np


df = pd.read_csv(file_path)

# Define mapping dictionary for payment methods
payment_mapping = {
    'upi': 'UPI',
    'phonepe': 'UPI',
    'googlepay': 'UPI',
    'bhim': 'UPI',

    'credit card': 'Credit Card',
    'creditcard': 'Credit Card',
    'cc': 'Credit Card',
    'visa': 'Credit Card',
    'mastercard': 'Credit Card',
    'BNPL' : 'Credit Card',

    'cash on delivery': 'Cash on Delivery',
    'cod': 'Cash on Delivery',
    'c.o.d': 'Cash on Delivery',
    'cash': 'Cash on Delivery',

    'debit card': 'Debit Card',
    'dc': 'Debit Card',
    'debitcard': 'Debit Card',

    'net banking': 'Net Banking',
    'netbanking': 'Net Banking',
    'nb': 'Net Banking',

    'emi': 'EMI',
    'installment': 'EMI',

    'wallet': 'Wallet',
    'paytm': 'Wallet',
    'amazonpay': 'Wallet',
    'mobikwik': 'Wallet'
}

# Function to clean payment method
def clean_payment_method(value):
    if pd.isna(value):
        return 'Unknown'
    value = str(value).lower().strip()
    # Check for exact key match first for a cleaner search in case of full strings
    if value in payment_mapping: 
        return payment_mapping[value]
    # Check for substring match (as in the original logic for cases like 'visa' in 'paid by visa cc')
    for key in payment_mapping:
        if key in value:
            return payment_mapping[key]
    return 'Other'

# Apply cleaning function
# Assuming the original column name is 'payment_method'
# If the column name is different, adjust 'payment_method' accordingly
df['payment_method'] = df['payment_method'].apply(clean_payment_method)

# Save the updated DataFrame back to the same file path string
df.to_csv(file_path, index=False)

print("✅ Payment method column cleaned and file updated successfully.")
```
